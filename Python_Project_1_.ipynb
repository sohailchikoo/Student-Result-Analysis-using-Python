**GENDER DISTRIBUTION**

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart we have analyzed that females are more then men .
# Females = 15424
# Males = 15217
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parents Education and Students Marks")
plt.show()
```

```python
# from the above we have concluded that the education of parents have good marks .
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parents Marital Status ")
plt.show()
```
